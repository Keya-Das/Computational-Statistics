```python
# Draw a line chart(static values)

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,15,12,18,20]

plt.plot(x,y,marker='*')
plt.title("LINE CHART")
plt.xlabel("values of x")
plt.ylabel("values of y")
plt.show()
```

```python
#Draw a line chart(User-Input Values)

import matplotlib.pyplot as plt
x_values = []
y_values = []
for i in range (5):
  x = int(input("Enter value of x : "))
  y = int(input("Enter value of y : "))
  x_values.append(x)
  y_values.append(y)

plt.plot(x_values, y_values, marker = '*')
plt.title("LINE CHART")
plt.xlabel("values of x")
plt.ylabel("values of y")
plt.show()
```

```python
# Draw a semi-logarithmic line chart (static values)

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,100,1000,10000,100000]

plt.semilogy(x, y, marker='*')

plt.title("semi-log line chart")
plt.xlabel("x-axis")
plt.ylabel("log y-axis")
plt.show()
```

```python
# Draw a logarithmic line chart (static values)

import matplotlib.pyplot as plt

x = [1,10,100,1000,10000]
y = [2,20,200,2000,20000]

#log-log plot (both axes logarithmic)
plt.loglog(x,y,marker='*')

plt.title("LOGARITHMIC LINE CHART")
plt.xlabel("log x-axis")
plt.ylabel("log y-axis")
plt.show()
```

```python
# Draw a band chart (static values)

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
lower = [10,12,15,13,17]
upper = [15,18,20,13,22]

plt.plot(x,lower)
plt.plot(x, upper)
plt.fill_between(x,lower,upper,alpha=0.5)

plt.title("BAND CHART")
plt.xlabel("x-axis")
plt.ylabel("values of y")
plt.show()
```

```python
# Draw a band chart (USER-INPUT)

import matplotlib.pyplot as plt
x = []
lower = []
upper = []
for i in range (5):
  x_values=int(input("Enter Value of x : "))
  y_lower=int(input("Enter lower Value of y : "))
  y_upper=int(input("Enter upper Value of y : "))
  x.append(x_values)
  lower.append(y_lower)
  upper.append(y_upper)

plt.plot(x,lower)
plt.plot(x,upper)
plt.fill_between(x,lower,upper,alpha=0.5)

plt.title("BAND CHART")
plt.xlabel("x-axis")
plt.ylabel("values of y")
plt.show
```

```python
# Draw a multiple bar chart (static values)

import matplotlib.pyplot as plt
import numpy as np

labels = ['A','B','C','D']
data1 = [10,20,15,20]
data2 = [12,18,17,22]

pos = np.arange(len(labels))
width = 0.3

plt.bar(pos - width/2,data1,width,label='data 1')
plt.bar(pos + width/2,data2,width,label='data 2')

plt.title("Multiple bar chart")
plt.xlabel("Labels")
plt.ylabel("values")
plt.xticks(pos,labels)
plt.legend()
plt.show()

```

```python
# Draw a multiple bar chart (USER-INPUT)

import matplotlib.pyplot as plt
import numpy as np
labels = []
data1 = []
data2 = []
n = int(input("Enter number of categorirs: "))

for i in range(n):
  x = input("Enter label name: ")
  y1 = int(input("Enter value for data 1: "))
  y2 = int(input("Enter value for data 2: "))
  labels.append(x)
  data1.append(y1)
  data2.append(y2)

pos = np.arange(len(labels))
width = 0.3

plt.bar(pos - width/2,data1,width,label='data 1')
plt.bar(pos + width/2,data2,width,label='data 2')

plt.title("Multiple Bar chart")
plt.xlabel("Labels")
plt.ylabel("values")
plt.xticks(pos,labels)
plt.legend()
plt.show()
```

```python
# Draw a composite/stacked bar chart (static values)

import matplotlib.pyplot as plt
labels = ['A','B','C','D']
data1 = [10,20,15,25]
data2 = [5,10,8,12]

plt.bar(labels,data1,label="part 1")
plt.bar(labels,data2,bottom=data1,label="part 2")
plt.title("composite bar chart")
plt.xlabel("labels")
plt.ylabel("values")
plt.legend()
plt.show()
```

```python
# Draw a composite/stacked bar chart (USER_INPUT)

import matplotlib.pyplot as plt
import numpy as np
labels = []
data1 = []
data2 = []
n = int(input("Enter number of categories: "))

for i in range(n):
  x = input("Enter label name: ")
  y1 = int(input("Enter value for Data 1: "))
  y2 = int(input("Enter value for Data 2: "))
  labels.append(x)
  data1.append(y1)
  data2.append(y2)


plt.bar(labels,data1,label="part 1")
plt.bar(labels,data2,bottom=data1,label="part 2")

plt.title("composite bar chart")
plt.xlabel("labels")
plt.ylabel("values")
plt.legend()
plt.show()
```

```python
# Draw a pie chart (static values)

import matplotlib.pyplot as plt

names = ['A','B','C','D']
sizes = [20,30,25,25]

plt.pie(sizes,labels=names,autopct='%1.1f%%')

plt.title("pie chart")
plt.show()
```

```python
#Draw a pie chart (USER_INPUT)

import matplotlib.pyplot as plt
names = []
sizes = []
n = int(input("Enter number of categories: "))

for i in range(n):
  name = input("Enter category name: ")
  value = float(input("Enter values: "))
  names.append(value)

  sizes.append(value)

plt.pie(sizes, labels=names,autopct='%1.1f%%')

plt.title("pie chart")
plt.show()
```
