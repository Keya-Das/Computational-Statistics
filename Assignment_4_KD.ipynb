```python
# Consider a set of discrete values. Find the mean deviation about any constant A and about mean

n = int(input("Enter number of values: "))
data = []
for i in range(n):
    value = float(input("Enter value: "))
    data.append(value)

A = float(input("Enter value of constant A: "))

mean = sum(data) / n

md_A = 0
for x in data:
    md_A += abs(x - A)
md_A = md_A / n

md_mean = 0
for x in data:
    md_mean += abs(x - mean)
md_mean = md_mean / n

print("\nMean =", mean)
print("Mean Deviation about A =", md_A)
print("Mean Deviation about Mean =", md_mean)
```

```python
# Consider a set of continuous values. Find the mean deviation about any constant A and about mean

import numpy as np
n = int(input("Enter number of classes: "))
lower = []
upper = []
freq = []
for i in range(n):
    l = float(input("Enter lower limit: "))
    u = float(input("Enter upper limit: "))
    f = float(input("Enter frequency: "))
    lower.append(l)
    upper.append(u)
    freq.append(f)

mid = []
for i in range(n):
    m = (lower[i] + upper[i]) / 2
    mid.append(m)

N = sum(freq)
mean = sum(f*m for f, m in zip(freq, mid)) / N
A = float(input("Enter value of constant A: "))

md_A = sum(f * abs(m - A) for f, m in zip(freq, mid)) / N
md_mean = sum(f * abs(m - mean) for f, m in zip(freq, mid)) / N

print("\nMean =", mean)
print("Mean Deviation about A =", md_A)
print("Mean Deviation about Mean =", md_mean)
```

```python
# Consider a set of discrete values. Find the standard deviation and about variance

import math
n = int(input("Enter number of values: "))
data = []
for i in range(n):
    value = float(input("Enter value: "))
    data.append(value)

mean = sum(data) / n

variance = 0
for x in data:
    variance += (x - mean) ** 2
variance = variance / n

std_dev = math.sqrt(variance)

print("\nMean =", mean)
print("Variance =", variance)
print("Standard Deviation =", std_dev)
```

```python
# Consider a set of continuous values. Find the standard deviation and about variance

import math
n = int(input("Enter number of classes: "))
lower = []
upper = []
freq = []
for i in range(n):
    l = float(input("Enter lower limit: "))
    u = float(input("Enter upper limit: "))
    f = float(input("Enter frequency: "))
    lower.append(l)
    upper.append(u)
    freq.append(f)

mid = []
for i in range(n):
    m = (lower[i] + upper[i]) / 2
    mid.append(m)

N = sum(freq)
mean = sum(f*m for f, m in zip(freq, mid)) / N

variance = 0
for f, m in zip(freq, mid):
    variance += f * (m - mean) ** 2
variance = variance / N

std_dev = math.sqrt(variance)

print("\nMean =", mean)
print("Variance =", variance)
print("Standard Deviation =", std_dev)
```
