```python
# Consider a set of discrete values. Find the first, second, third and fourth raw moments

n = int(input("Enter number of values: "))
data = []
for i in range(n):
    value = float(input("Enter value: "))
    data.append(value)

m1 = sum(x for x in data) / n
m2 = sum(x**2 for x in data) / n
m3 = sum(x**3 for x in data) / n
m4 = sum(x**4 for x in data) / n

print("\nFirst Raw Moment =", m1)
print("Second Raw Moment =", m2)
print("Third Raw Moment =", m3)
print("Fourth Raw Moment =", m4)
```

```python
# Consider a set of continuous values. Find the first, second, third and fourth raw moments

n = int(input("Enter number of classes: "))
lower = []
upper = []
freq = []

for i in range(n):
    l = float(input("Enter lower limit: "))
    u = float(input("Enter upper limit: "))
    f = float(input("Enter frequency: "))
    lower.append(l)
    upper.append(u)
    freq.append(f)

mid = []
for i in range(n):
    m = (lower[i] + upper[i]) / 2
    mid.append(m)

N = sum(freq)

m1 = sum(f*m for f, m in zip(freq, mid)) / N
m2 = sum(f*(m**2) for f, m in zip(freq, mid)) / N
m3 = sum(f*(m**3) for f, m in zip(freq, mid)) / N
m4 = sum(f*(m**4) for f, m in zip(freq, mid)) / N

print("\nFirst Raw Moment =", m1)
print("Second Raw Moment =", m2)
print("Third Raw Moment =", m3)
print("Fourth Raw Moment =", m4)
```

```python
# Consider a set of discrete values. Find the first, second, third and fourth central moments

n = int(input("Enter number of values: "))
data = []
for i in range(n):
    value = float(input("Enter value: "))
    data.append(value)

mean = sum(data) / n

m1 = sum((x - mean) for x in data) / n
m2 = sum((x - mean)**2 for x in data) / n
m3 = sum((x - mean)**3 for x in data) / n
m4 = sum((x - mean)**4 for x in data) / n

print("\nMean =", mean)
print("First Central Moment =", m1)
print("Second Central Moment =", m2)
print("Third Central Moment =", m3)
print("Fourth Central Moment =", m4)
```

```python
# Consider a set of continuous values. Find the first, second, third and fourth central moments

n = int(input("Enter number of classes: "))
lower = []
upper = []
freq = []

for i in range(n):
    l = float(input("Enter lower limit: "))
    u = float(input("Enter upper limit: "))
    f = float(input("Enter frequency: "))
    lower.append(l)
    upper.append(u)
    freq.append(f)

mid = []
for i in range(n):
    m = (lower[i] + upper[i]) / 2
    mid.append(m)

N = sum(freq)
mean = sum(f*m for f, m in zip(freq, mid)) / N

m1 = sum(f*(m - mean) for f, m in zip(freq, mid)) / N
m2 = sum(f*(m - mean)**2 for f, m in zip(freq, mid)) / N
m3 = sum(f*(m - mean)**3 for f, m in zip(freq, mid)) / N
m4 = sum(f*(m - mean)**4 for f, m in zip(freq, mid)) / N

print("\nMean =", mean)
print("First Central Moment =", m1)
print("Second Central Moment =", m2)
print("Third Central Moment =", m3)
print("Fourth Central Moment =", m4)
```

```python
# Find the measures of skewness ( Karl Pearson )

import math
n = int(input("Enter number of values: "))
data = []
for i in range(n):
    value = float(input("Enter value: "))
    data.append(value)

mean = sum(data) / n

data.sort()
if n % 2 == 0:
    median = (data[n//2 - 1] + data[n//2]) / 2
else:
    median = data[n//2]

variance = sum((x - mean)**2 for x in data) / n
sd = math.sqrt(variance)

sk = 3 * (mean - median) / sd

print("\nMean =", mean)
print("Median =", median)
print("Standard Deviation =", sd)
print("Skewness =", sk)
```

```python
# Find the measure of kurtosis

n = int(input("Enter number of values: "))
data = []
for i in range(n):
    value = float(input("Enter value: "))
    data.append(value)
mean = sum(data) / n

m2 = sum((x - mean)**2 for x in data) / n
m4 = sum((x - mean)**4 for x in data) / n

beta2 = m4 / (m2**2)
gamma2 = beta2 - 3

print("\nKurtosis (β2) =", beta2)
print("Excess Kurtosis (γ2) =", gamma2)

if beta2 == 3:
    print("Mesokurtic")
elif beta2 > 3:
    print("Leptokurtic")
else:
    print("Platykurtic")
```
