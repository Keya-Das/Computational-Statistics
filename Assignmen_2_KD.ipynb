```python
# Find frequency, relative frequency, cumulative frequency (less-than type and greater than type) from a set of discrete data.Draw a column diagram for a set of discrete data

import matplotlib.pyplot as plt

data = [2, 3, 2, 5, 3, 2, 4, 5, 3, 2]
# Frequency
freq = {}
for i in data:
    freq[i] = freq.get(i, 0) + 1
x = sorted(freq.keys())
f = [freq[i] for i in x]

# Relative Frequency
total = sum(f)
rel_freq = [i/total for i in f]

# Cumulative Frequency (Less-than)
cum_less = []
s = 0
for i in f:
    s += i
    cum_less.append(s)

# Cumulative Frequency (Greater-than)
cum_greater = []
s = total
for i in f:
    cum_greater.append(s)
    s -= i

print("Value   Freq   RelFreq   Cum<   Cum>")
for i in range(len(x)):
    print(x[i], "   ", f[i], "   ", round(rel_freq[i],2),
          "   ", cum_less[i], "   ", cum_greater[i])

plt.bar(x, f)
plt.title("Column Diagram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Draw a cumulative frequency diagram (less-than type)  and greater-than type) for a set of discrete data. (STATIC VALUES)

import matplotlib.pyplot as plt

values = [1, 2, 3, 4, 5]
cum_less = [1, 3, 6, 8, 9]
cum_greater = [9, 8, 6, 3, 1]

plt.plot(values, cum_less, marker="*", label="Less Than")
plt.plot(values, cum_greater, marker="*", label="Greater Than")

plt.title("Cumulative Frequency Curve")
plt.xlabel("Values")
plt.ylabel("Cumulative Frequency")
plt.legend()
plt.show()
```

```python
# Find frequency, relative frequency, cumulative frequency (less-than type and greater than type) from a set of continuous data.
# USER-INPUT

import matplotlib.pyplot as plt
import numpy as np

n = int(input("Enter number of data values: "))
data = []
for i in range(n):
    value = float(input("Enter value: "))
    data.append(value)

k = int(input("Enter number of classes: "))
bins = []
for i in range(k + 1):
    b = float(input("Enter class boundary: "))
    bins.append(b)

freq, edges = np.histogram(data, bins)

total = sum(freq)
rel_freq = []
for i in freq:
    rel_freq.append(i / total)

cum_less = []
s = 0
for i in freq:
    s += i
    cum_less.append(s)

cum_greater = []
s = total
for i in freq:
    cum_greater.append(s)
    s -= i

print("\nClass Interval   Freq   RelFreq   Cum<   Cum>")
for i in range(len(freq)):
    interval = f"{edges[i]}-{edges[i+1]}"
    print(f"{interval:<15} {freq[i]:<6} {round(rel_freq[i],2):<8} {cum_less[i]:<5} {cum_greater[i]:<5}")
```

```python
# Draw a less-than type ogive for a set of continuous data(STATIC VALUES)

import matplotlib.pyplot as plt
import numpy as np

data = [12, 15, 17, 19, 21, 23, 25, 27, 29]
bins = [10, 15, 20, 25, 30]

freq, edges = np.histogram(data, bins)
cum_freq = np.cumsum(freq)
upper_limits = edges[1:]

plt.plot(upper_limits, cum_freq, marker="*")
plt.title("Less Than Ogive")
plt.xlabel("Upper Class Boundary")
plt.ylabel("Cumulative Frequency")
plt.show()
```

```python
# Draw a greater-than type ogive for a set of continuous data.(STATIC VALUES)

import matplotlib.pyplot as plt
import numpy as np

data = [12, 15, 17, 19, 21, 23, 25, 27, 29]
bins = [10, 15, 20, 25, 30]

freq, edges = np.histogram(data, bins)
cum_greater = np.cumsum(freq[::-1])[::-1]
lower_limits = edges[:-1]

plt.plot(lower_limits, cum_greater, marker="*")
plt.title("Greater Than Ogive")
plt.xlabel("Lower Class Boundary")
plt.ylabel("Cumulative Frequency")
plt.show()
```
